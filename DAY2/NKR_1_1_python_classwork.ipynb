#### Boolean indexing

```python
bool_a=np.array([[1,2],[3,4],[5,6]],)
bool_indx=(bool_a>2)
bool_indx
```

```python
bool_a[bool_indx]
```

```python
bool_b=np.array(['python','java','ruby','rust'])
bool_indx=((bool_b == 'python') & (bool_b == 'ruby'))
bool_indx
```

```python
a=np.arange(40)
a[a%2==0]
```

#### Matrix addition and subtraction

```python
## single element -1 is added to all the each element of the x
x=np.array([1,2,3])
y=np.array([-1])
output=np.add(x,y)
output
```

```python
## This approache is not possible because of the shape conflict
x=np.array([1,2,3])
y=np.array([-1,2])
output=np.add(x,y)
output
```

## Statistics and Linear Algebra

```python
## Transpose
x=np.array([[1,2,3,4],[5,6,7,8]])
x.T
```

```python
sum_a=x.sum()
sum_a
```

```python
## COLUMN WISE SUM
x.sum(axis=0)
```

```python
## ROW WISE SUM
x.sum(axis=1)
```

#### Mean

```python
x.mean()
```

```python
x.mean(axis=1)
```

```python
x.mean(axis=0)
```
